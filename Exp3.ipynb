# 필요한 모듈 import 하기

```python
import cv2    # 이미지 처리를 위한 openCV
import matplotlib.pyplot as plt    # 노트북에 이미지를 출력하기 위한 matplotlib
import numpy as np
import dlib
import os
```

# 필요한 이미지 불러오기

```python
my_image_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/image1.jpg'
img_bgr = cv2.imread(my_image_path)
img_bgr_shape = img_bgr.shape
img_bgr = cv2.resize(img_bgr, (img_bgr_shape[1]//4,img_bgr_shape[0]//4))    # Resize
img_show = img_bgr.copy()    # 출력용 이미지 별도 보관
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)     # BGR에서 RGB로 변환
plt.imshow(img_rgb)
plt.show()
```

# 얼굴 검출하기

```python
detector_hog = dlib.get_frontal_face_detector()    # dlib 정면 얼굴 검출기로 얼굴 검출 후 반환
dlib_rects = detector_hog(img_rgb, 1)    # (이미지, 이미지 피라미드 수) 이미지 피라미드수:이미지를 unsampling하여 크기를 키움
```

```python
print(dlib_rects)   # 찾은 얼굴영역 좌표

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)    # 초록색, 두께:2로 얼굴 위치 표현 

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

# 랜드마크 검출하기

```python
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)    # 이미지의 얼굴에서 랜드마크 찾기

list_landmarks = []
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

for landmark in list_landmarks:
    for idx, point in enumerate(list_points):
        cv2.circle(img_show, point, 2, (0, 255, 255), -1)
        # cv2.circle(img, center, radian, color, thickness), 선의 두께가 -1이면 원 안쪽
        
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

# 하트스티커가 적용될 위치 지정

```python
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print('코 위치: ', landmark[30])
    # 수염 크기 조정
    w = (landmark[30][0] - landmark[36][0])*2 
    h = (landmark[48][1] - landmark[36][1])*2
    # 수염 시작 위치 지정
    x = landmark[30][0] - w//2
    y = landmark[30][1] - h//2
    print ('스티커 위치 (x,y) : (%d,%d)'%(x,y))
    print ('스티커 크기 (w,h) : (%d,%d)'%(w,h))
```

# 스티커 불러오기

```python
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat1.png'
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w, h))    # 설정한 스티커 크기로 리사이즈
print ('이미지 스티커 shape: ', img_sticker.shape)
```

# 스티커 적용

```python
sticker_area = img_show[y:y+img_sticker.shape[0], x:x+img_sticker.shape[1]]

img_show[y:y+img_sticker.shape[0],  x:x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)

plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

# 실제 사진에 적용하기

```python
sticker_area = img_bgr[y:y+img_sticker.shape[0], x:x+img_sticker.shape[1]]
img_bgr[y:y+img_sticker.shape[0],  x:x+img_sticker.shape[1]] = \
    np.where(img_sticker==255, sticker_area,img_sticker).astype(np.uint8)

plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```

# 랜드마크 검출하기

```python
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)    # 이미지의 얼굴에서 랜드마크 찾기

list_landmarks = []
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

for landmark in list_landmarks:
    for idx, point in enumerate(list_points):
        cv2.circle(img_show, point, 2, (0, 255, 255), -1)
        # cv2.circle(img, center, radian, color, thickness), 선의 두께가 -1이면 원 안쪽
        
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

# 스티커 불러오기

```python
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat1.png'
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w, h))    # 설정한 스티커 크기로 리사이즈
print ('이미지 스티커 shape: ', img_sticker.shape)
```

# 회전된 사진 각도 구하기

```python
import math

# 아크탄젠트로 기울기 구하기
# 양쪽 눈끝을 기준으로 각도 구함
dx = landmark[45][0] - landmark[36][0]
dy = landmark[45][1] - landmark[36][1]

angle = abs(math.atan(dy/dx)/math.pi*180)

if dy>0:    # y좌표는 위에 있을수록 값이 작으므로 두 좌표의 차가 0보다 크면 오른쪽으로 사진이 기울어졌다고 판단
    angle = -angle
print('각도: ', angle)
```

# affine 함수로 사진 회전시키기

```python
height, width = img_sticker.shape[:2]
m = cv2.getRotationMatrix2D((width/2, height/2), angle, 1)
print(m)
img_sticker = cv2.warpAffine(img_sticker, m, (width, height), borderMode=cv2.BORDER_DEFAULT)
```

# 스티커 적용

```python
sticker_area = img_show[y:y+img_sticker.shape[0], x:x+img_sticker.shape[1]]

img_show[y:y+img_sticker.shape[0],  x:x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)

plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
sticker_area = img_bgr[y:y+img_sticker.shape[0], x:x+img_sticker.shape[1]]
img_bgr[y:y+img_sticker.shape[0],  x:x+img_sticker.shape[1]] = \
    np.where(img_sticker==255, sticker_area,img_sticker).astype(np.uint8)

plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```

검출된 랜드마크에서 설정한 스티커사진 크기를 가지고 회전된 각도를 구했다.   
처음에는 구글에서 다른 방법 찾아서 각도를 구했는데 사진에 잘 적용이 안됐다.   
그래서 그냥 양쪽 눈 사이를 끝점으로 해서 각도를 구했다.   
그다음 cv.rotate를 사용해서 회전 시키려고 했는데 90, 180 이렇게밖에 안되는 것 같았다.   
그래서 행렬을 만들고 전환시키는 방법으로 스티커를 얼굴 각도에 맞게 회전시켰다.   
warpAffine 함수에서 borderMode=transparent로 설정했더니 테두리가 검은게 부분 남았다.   
찾아봤는데 이유는 모르겠고 BORDER_DEFAULT로 했더니 원래 사진과 합성돼서 잘 나왔다.   
수염 회전 후 화질이 떨어지는데 어떻게 올려야할지 모르겠다.
